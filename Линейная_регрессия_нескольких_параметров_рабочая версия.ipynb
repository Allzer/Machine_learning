```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

import seaborn as sns
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm #линейная регрессия, но с большим кол-вом статистических параметров
```

Загружаем таблцу

```python
f = pd.read_csv('student-mat-2.csv', encoding="latin-1", sep=",")
```

Занесём результаты экзов в отдельные массивы

```python
obrM = f['ObrM'][:]
work_a = f['workDay_alko'][:]
ex1 = f['ex1'][:]
obrP = f['ObrP'][:]
timePoezdk = f['timePoezdk'][:] # f[''][]
timeYcheb = f['timeYcheb'][:]
Otin_in_family = f['Otin_in_family'][:]
freeTime = f['freeTime'][:]
friendTime = f['friendTime'][:]
workDay_alko = f['workDay_alko'][:]
freeDay_walk = f['freeDay_walk'][:]
zdorov = f['zdorov'][:]
propusk = f['propusk'][:]
ex2 = f['ex2'][:]
ex3 = f['ex3'][:]
```

```python
colvo = 0
for i in range(len(ex1)):
  if ex1[i]==8:
    colvo = colvo + 1
print('Колличество оценки 10 = ', colvo)
```

Визуализируем

```python
plt.figure(figsize=(10,6))
plt.hist(ex1, bins=50, ec='black',color='b')

plt.xlabel('Оценка за экзамен')
plt.ylabel('Кол-во учеников')
plt.show()
```

```python
new = f.drop('ex1',axis=1)
new
```

Обучаем модель, также будем отсеивать данные

```python
x_train, x_test, y_train, y_test = train_test_split(new, ex1, test_size=0.2, random_state = 10) #Распределяем данные для тренировки и для тестов

model = LinearRegression().fit(x_train,y_train)
print('Коэффициент детерминации: ', model.score(x_train,y_train)) #Коэффициент детерминации показывает точность модели
print('Коэффициент детерминации тестов: ', model.score(x_test,y_test))

y_train = np.array(y_train)
print(y_train)
```

Получаем прогноз

```python
oc_pred = model.predict(x_train)
oc_pred[:] #запредиктил оценки всех учеников
```
