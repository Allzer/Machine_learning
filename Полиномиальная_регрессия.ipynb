```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from pylab import *
from sklearn.metrics import r2_score
```

```python
f = pd.read_csv('kc_house_data.csv')
f
```

```python
f.drop('id',axis=1)
```

```python
s_l = f['sqft_living'][::100] #Берём не все значения т.к график будет ущербным
price = f['price'][::100]
```

```python
scatter(s_l,price,s=1)
```

```python
x = np.array(s_l)
y = np.array(price)
```

```python
p5 = np.poly1d(np.polyfit(x,y,4)) #Решаем полнином четвёртой степени
xp = np.linspace(0,6700,8000) #Задаём параметры для линии, которая описывает наш полном
plt.scatter(x,y,s=10) #Рисуем точки на граффике
plt.plot(xp,p5(xp),c='r') #Рисуем линию полинома
plt.show
```
